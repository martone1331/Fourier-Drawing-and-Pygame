```python
import pygame
import math
import sys
import numpy
import os
import random

#define screen resolution and origin
res_w, res_h = 1920, 1080
origin_x = res_w//2
origin_y = res_h//2

#initialize pygame screen and clock
pygame.init()
pygame.font.init()
screen = pygame.display.set_mode((res_w, res_h))
font = pygame.font.SysFont('courier', 40)
font2 = pygame.font.SysFont('courier', 20)

pygame.display.set_caption('Fourier Drawing')
clock = pygame.time.Clock()


def DFT(x):
    out = []
    N= len(x)
    for k in range(1,N):
        real = 0
        im = 0
        for n in range(N):
            real += x[n]*math.cos((2*math.pi*k*n/N))
            im -= x[n]*math.sin((2*math.pi*k*n/N))
        
        real = real/N
        im = im/N
        freq = k
        amp = math.sqrt(real**2 + im**2)
        out.append([freq,amp])
    return out

#parameters
time = 0
draw_x= []
draw_y = []

x = []
y = []

for k in range(0, 360, 1):
    x_ang = k*(math.pi/180)
    x.append(math.cos(x_ang))
        
for k in range(0, 360, 1):
    y_ang = k*(math.pi/180)
    y.append(math.sin(y_ang))


my_input_x = DFT(x)
my_input_y = DFT(y)
print(my_input_x)
print(my_input_y)

def circles(offset_x, offset_y, transform, phase):
    center_x = origin_x - offset_x
    center_y = origin_y - offset_y
    
    for i in range(0, len(transform)):
        scale = 100
            
        amp = scale*float(transform[i][1])
        freq = float(transform[i][0])
        
        dx = amp*math.cos(freq*time+phase)
        dy = amp*math.sin(freq*time+phase)
    
    
    #drawing circles
        circle = pygame.draw.circle(screen, 'White', (center_x,center_y), amp, width = 1,)
        circle_pointer = pygame.draw.circle(screen, 'White', (center_x-dx,center_y+dy), 8, width = 1)
        radial = pygame.draw.line(screen, 'White', (center_x,center_y), (center_x-dx,center_y+dy))
        
        center_x = center_x-dx
        center_y = center_y+dy
    
    return center_x, center_y
            

#all shapes/objects rendered here
def render():
    screen.fill('Black') 
    
    #drawing the curve
    center_x_x, center_x_y = circles(300, -300, my_input_x, 0)
    center_y_x, center_y_y = circles(-300, 300, my_input_y, math.pi/2)

    draw_x.insert(0, (center_x_y))
    draw_y.insert(0, (center_y_y))
                  
    for l in range(0, len(draw_x)):
         graph = pygame.draw.circle(screen, 'Yellow', (draw_x[l], draw_y[l]), 1, width = 1)
            
    if len(draw_x) > 750:
         draw_x.pop()
            
    connect = pygame.draw.line(screen, 'Grey', (center_x_x,center_x_y), (origin_x,center_x_y))
    connect = pygame.draw.line(screen, 'Grey', (center_y_x,center_y_y), (origin_x,center_y_y))

        

#updater
RUNNING, PAUSE = 1, 0
state = RUNNING
wave = 'square'
degree = 1

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                if state == RUNNING:
                    state = PAUSE
                elif state == PAUSE:
                    state = RUNNING
            if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    sys.exit()
                
    else:
        if state == RUNNING:
            render()
            time+= (1/60)

    #set so timescale is 1 second
    
    
    pygame.display.update()
    
    #framerate
    clock.tick(60)

```
