```python
import pygame
import math
import sys
import numpy
import os
import random

def DFT(x):
    out = []
    N= len(x)
    for k in range(N):
        real = 0
        im = 0
        for n in range(N):
            real += x[n]*math.cos((2*math.pi*k*n/N))
            im -= x[n]*math.sin((2*math.pi*k*n/N))
        
        freq = 2*math.pi*k*n/N
        amp = math.sqrt(real**2 + im**2)
        out.append([freq,amp])
    return out

#define screen resolution and origin
res_w, res_h = 1920, 1080
origin_x = res_w//2
origin_y = res_h//2

#initialize pygame screen and clock
pygame.init()
pygame.font.init()
screen = pygame.display.set_mode((res_w, res_h))
font = pygame.font.SysFont('courier', 40)
font2 = pygame.font.SysFont('courier', 20)

pygame.display.set_caption('Fourier Drawing')
clock = pygame.time.Clock()


#parameters
time = 0
radius = 150
draw = []

example = []
for x in range(100):
    example.append(x)

wave = DFT(example)
print(wave)

#all shapes/objects rendered here
def render(wave_type):
    screen.fill('Black') 
   
    center_x = origin_x-400
    center_y = origin_y
    
    for i in range(0, len(wave)):
        
        
        if wave_type == 'square':
            scale = 2
            
            amp = scale*float(wave[i][1])
            freq = float(wave[i][0])
        
            dx = amp*math.cos(freq*time)
            dy = amp*math.sin(freq*time)
    
    
    #drawing circles
        circle = pygame.draw.circle(screen, 'White', (center_x,center_y), amp, width = 1,)
        circle_pointer = pygame.draw.circle(screen, 'White', (center_x-dx,center_y+dy), 8, width = 1)
        radial = pygame.draw.line(screen, 'White', (center_x,center_y), (center_x-dx,center_y+dy))
        
        center_x = center_x-dx
        center_y = center_y+dy
        
    #drawing the curve
        
        if i == len(wave)-1:
            draw.insert(0,(center_y))
            
            for l in range(0, len(draw)):
                graph = pygame.draw.circle(screen, 'Yellow', (origin_x+l,draw[l]), 1, width = 1)
            
            if len(draw) > 750:
                draw.pop()
            
            connect = pygame.draw.line(screen, 'Grey', (center_x,center_y), (origin_x,center_y))
        

#updater
RUNNING, PAUSE = 1, 0
state = RUNNING
wave = 'square'
degree = 1

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                if state == RUNNING:
                    state = PAUSE
                elif state == PAUSE:
                    state = RUNNING
            if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    sys.exit()
                
    else:
        if state == RUNNING:
            render(wave)
            time+= (1/60)

   
    #set so timescale is 1 second
    
    
    pygame.display.update()
    
    #framerate
    clock.tick(60)

def resource_path(relative_path):
    try:
    # PyInstaller creates a temp folder and stores path in _MEIPASS
        base_path = sys._MEIPASS
    except Exception:
        base_path = os.path.abspath(".")

    return os.path.join(base_path, relative_path)

```
